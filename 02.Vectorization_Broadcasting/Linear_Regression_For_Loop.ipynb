```python
# Making the imports
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0, 9.0)

# Preprocessing Input data
data = pd.read_csv('../01.Linear_Regression_with_Gradient_Descent/data.csv')
X = data.iloc[:, 0]
Y = data.iloc[:, 1]
plt.scatter(X, Y)
plt.show()
```

```python
print(Y[:5])
print(X[:5])
```

```python
# Getting input size
print(X.shape)
n = X.shape[0]
n
```

```python
%%time
# Calculating partial derivatives with respect to m and c from cost function
# Doing Gradient Descent
m = 1
c = 0
L = 0.0001  # The learning Rate

epochs = 10000  # The number of iterations to perform gradient descent

for _ in range(epochs):
    D_m = 0
    D_c = 0
    D_m_sum = 0
    D_c_sum = 0
    for i in range(n):
        y_hat = m*X[i] + c
        D_m += X[i]*(Y[i] - y_hat)
        D_c += Y[i] - y_hat
    D_m_sum = (-2/n)*D_m
    D_c_sum = (-2/n)*D_c
    m = m - L*D_m_sum
    c = c - L*D_c_sum

print(m)
print(c)
```

```python
%%time
# hypothesis for all the samples For Loop
Y_pred = []

for i in range(n):
    y_hat = m*X[i] + c
    Y_pred.append(y_hat)

print(Y[:10])
print(Y_pred[:10])
```

```python
plt.rcParams['figure.figsize'] = (12.0, 9.0)
plt.scatter(X, Y)
plt.scatter(X, Y_pred)
plt.show()
```
