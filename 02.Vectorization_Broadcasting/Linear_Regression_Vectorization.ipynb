```python
# Making the imports
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0, 9.0)

# Preprocessing Input data
data = pd.read_csv('../01.Linear_Regression_with_Gradient_Descent/data.csv')
X = data.iloc[:, 0]
Y = data.iloc[:, 1]
plt.scatter(X, Y)
plt.show()
```

```python
# Getting input size
print(X.shape)
n = X.shape[0]
n
```

```python
%%time

m = 1
c = 0

L = 0.0001  # The learning Rate

epochs = 10000  # The number of iterations to perform gradient descent

for _ in range(epochs):
    Y_pred = m*X + c  # The current predicted value of Y
    D_m = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt m
    D_c = (-2/n) * sum(Y - Y_pred)  # Derivative wrt c
    m = m - L * D_m  # Update m
    c = c - L * D_c  # Update c

print(c,"\n", m)
```

```python
%%time

m = 1
c = 0

theta = np.array([[c] ,[m]])

L = 0.0001  # The learning Rate

epochs = 10000  # The number of iterations to perform gradient descent

ones = np.ones((n, 1)) # column vector with ones
Z_x = X.values # converting from pandas series to numpy array
Z_x = Z_x.reshape((n, 1)) # reshaping X values to column vector
Z_x = np.append(ones, Z_x,  axis=1) # Adding ones to the left of X values
Z_y = Y.values # converting from pandas series to numpy array
Z_y = Z_y.reshape((n, 1)) # reshaping Y values to column vector

for _ in range(epochs):
    Y_pred = np.dot(Z_x, theta)  # The current predicted value of Y
    derivatives = (1/n)*np.transpose(Z_x)@(Y_pred - Z_y) # @-same as np.transpose
    theta = theta - L*derivatives # scalar vector multiplication

print(derivatives)
print(theta)
```
